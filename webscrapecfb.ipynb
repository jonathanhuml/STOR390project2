```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#url: https://www.sports-reference.com/cfb/schools/duke/2019-schedule.html
##div_schedule , .left , .right , .center
```

```python
page = requests.get("https://www.sports-reference.com/cfb/schools/duke/2019-schedule.html")
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.find(id = 'div_schedule'))
table = soup.find(id = 'div_schedule')
```

```python
thead = table.find('thead')
th_head = thead.find_all('th')
th_head
```

```python
names = []
for thh in th_head:
   names.append(thh.get('data-stat'))
names
```

```python
tbody = table.find('tbody')
tr_body = tbody.find_all('tr')
tr_body

a = []
for trb in tr_body:
    d = []
    th = trb.find('th')
    x = (th.get_text())
    d.append(x)
    a.append(d)
    
    for td in trb.find_all('td'):
        y = (td.get_text())
        d.append(y)
    

df = pd.DataFrame(a, columns =names)
df
```
