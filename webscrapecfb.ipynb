```python
#names must adhere to url profile of that team on sports reference
#I typed this because I only wanted a handful of teams and didn't always know what the name would be represented as
teams = ['duke', 'clemson', 'north-carolina', 'north-carolina-state', 'wake-forest',
        'syracuse', 'florida-state','georgia-tech','virginia-tech','louisville',
        'notre-dame','boston-college']

team_matrix = []
for names in range(len(teams)):
    page = requests.get("https://www.sports-reference.com/cfb/schools/" + teams[names] + "/2019-schedule.html")
    soup = BeautifulSoup(page.content, 'html.parser')
    table = soup.find(id = 'div_schedule')
    tbody = table.find('tbody')
    tr_body = tbody.find_all('tr')
    tr_body

    for trb in tr_body:
        column_matrix = []
        th = trb.find('th')
        th_text = (th.get_text())
        column_matrix.append(th_text)
        team_matrix.append(column_matrix)
    
        for td in trb.find_all('td'):
            td_text = (td.get_text())
            column_matrix.append(td_text)
    

team_dataframe = pd.DataFrame(team_matrix)
team_dataframe
```
