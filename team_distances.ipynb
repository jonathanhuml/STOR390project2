```python
wiki = requests.get("https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_FBS_football_stadiums")
wiki_content = BeautifulSoup(wiki.content, 'html.parser')
```

```python
names = []
for column_names in wiki_content.find_all('th'):
    names.append(column_names.get_text())
names
```

```python
table_body = wiki_content.find('tbody')
cells = table_body.find_all('tr')

data_matrix = []
for i in range(len(cells)):
    team_row = []
    for content in cells[i].find_all('td'):
        text = content.get_text()
        team_row.append(text)
    data_matrix.append(team_row)

df = pd.DataFrame(data_matrix)
df = df.replace('\n','', regex=True)
team_city = df[df.columns[2:5]]


predictions = pd.read_csv('/Users/juanhuml/Desktop/STOR390/Data/Predictions.csv')
home = pd.merge(predictions, team_city, left_on='Home Team', right_on=4)
home_away = pd.merge(home, team_city, left_on = 'Visitor Team', right_on = 4)
cities = home_away[home_away.columns[13:]]
cities
```
